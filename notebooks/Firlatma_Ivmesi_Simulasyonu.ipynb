# Ktle Frlatma vmesi ve Enerji Profili Simlasyonu
Bu notebook, Ay Güney Kutbu altyapsndaki elektromanyetik katapultun fiziksel simlasyonunu grselletirir.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_launch(mass, target_v, accel_g):
    a = accel_g * 9.81
    t_final = target_v / a
    t = np.linspace(0, t_final, 100)
    v = a * t
    d = 0.5 * a * (t**2)
    
    plt.figure(figsize=(10, 4))
    plt.plot(t, v, label='Hz (m/s)')
    plt.plot(t, d/10, label='Mesafe (dam)')
    plt.title('Frlatma Profili')
    plt.xlabel('Zaman (s)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
